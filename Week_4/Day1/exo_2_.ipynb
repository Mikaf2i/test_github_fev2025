```python
# -*- coding: utf-8 -*-
"""Perceptron_Exercice.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/YOUR_DRIVE_ID
"""

def perceptron(temperature, pluie):
    """
    Simule un perceptron pour décider si sortir.

    Args:
        temperature (float): Température en °F.
        pluie (int): 1 si pluie, 0 sinon.

    Returns:
        int: 1 si sortir, 0 si rester.
    """
    poids_temperature = 0.6
    poids_pluie = 0.4
    biais = 2

    somme_ponderee = (temperature * poids_temperature) + (pluie * poids_pluie) + biais

    if somme_ponderee > 20:
        return 1  # Oui, aller à l'extérieur
    else:
        return 0  # Non, rester à l'intérieur

# Cas 1: Température = 70°F, Pluie = 0 (Non)
temperature1 = 70
pluie1 = 0
resultat1 = perceptron(temperature1, pluie1)
print(f"Cas 1 : Température = {temperature1}°F, Pluie = {pluie1}")
print(f"Sortir ? : {'Oui' if resultat1 == 1 else 'Non'}")

# Cas 2: Température = 50°F, Pluie = 1 (Oui)
temperature2 = 50
pluie2 = 1
resultat2 = perceptron(temperature2, pluie2)
print(f"\nCas 2 : Température = {temperature2}°F, Pluie = {pluie2}")
print(f"Sortir ? : {'Oui' if resultat2 == 1 else 'Non'}")

# Interprétation des résultats
print("\nInterprétation des résultats :")
if resultat1 == 1 and resultat2 == 1:
    print("Le perceptron a suggéré de sortir dans les deux cas.")
else:
    print("Le perceptron n'a pas suggéré de sortir dans les deux cas.")

print("Dans les deux cas, la somme pondérée calculée était supérieure au seuil de 20.")
print("Par conséquent, la fonction d'activation a renvoyé 1, ce qui indique que la décision est de sortir.")
```
