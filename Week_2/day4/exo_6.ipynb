```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

Panic_attacks = {
    'Age' : [20, 30, 40, 50],
    'Do you have Panic Attacks?': ['Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(Panic_attacks)
print(df)

# Convert Yes/No to 1/0 for regression
df['attaque'] = df['attaque'].map({'Yes': 1, 'No': 0})

sns.lmplot(x='Age', y='attaque', data=df, aspect=1.5, height=5)
plt.title('la relation entre lâge et les crises de panique')
plt.show()

```
