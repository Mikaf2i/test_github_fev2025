```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Name': ['Tom', 'Nick', 'John', 'Alisa', 'Peter', 'Tom', 'Nick', 'John', 'Alisa', 'Peter'],
    'note': [3.5, 3.0, 3.5, 4.0, 3.5, 3.5, 3.0, 3.5, 4.0, 3.5],
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.histplot(df['note'], bins=5, kde=True, color="blue") 

# Ajout de titres
plt.title("Distribution des CGPA")
plt.xlabel("Valeur du CGPA")
plt.ylabel("Fr√©quence")
plt.show()
```
